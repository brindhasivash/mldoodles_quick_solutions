# Solution: Matplotlib Save Plot - Saves a Blank Plot

Please refer to the blog post https://mldoodles.com/matplotlib-saves-blank-plot/ for more information.

```python
import matplotlib.pyplot as plt
```

## Problem

```python
x = [1, 2, 3, 4, 5]
y = [4, 7, 5, 6, 3]
plt.bar(x, y)
plt.show()
plt.savefig("myfigure.png")
```

## Solution - 1

```python
x = [1, 2, 3, 4, 5]
y = [4, 7, 5, 6, 3]
plt.bar(x, y)
plt.title("Solution-1")
plt.savefig("myfigure2.png")
plt.show()
```

## Solution - 2

```python
x = [1, 2, 3, 4, 5]
y = [4, 7, 5, 6, 3]
fig, ax = plt.subplots()
ax.bar(x, y)
plt.title("Solution-2")
plt.show()
fig.savefig("myfigure3.png")
```
